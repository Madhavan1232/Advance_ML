```python
import pandas as pd 

url = "https://github.com/Madhavan1232/Advance_ML/raw/refs/heads/main/ClassWorks/Day3/p2/ML470_S3_Diabetes_Data_Preprocessed_Concept.xlsx"

df = pd.read_excel(url)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

X = df.drop(columns=[df.columns[-1]])
y = df[df.columns[-1]]

clf = DecisionTreeClassifier(criterion='entropy', max_depth=2, random_state=42)
clf.fit(X, y)

plt.figure(figsize=(14, 10))
plot_tree(clf, 
          feature_names=X.columns.tolist(), 
          filled=True, 
          rounded=True, 
          precision=3,
          fontsize=10)

plt.title("Decision Tree Visualization for Diabetes Risk Prediction")
plt.show()
```
