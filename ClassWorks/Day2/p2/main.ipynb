```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://github.com/LuckyBoy587/Advanced-ML-Techniques/raw/refs/heads/master/Day2/P2/ML470_S2_Practice_HR_Data.xlsx"
df = pd.read_excel(url)
```

```python
df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10, 7))
sns.boxplot(data=df)
plt.xticks(rotation=45)
```
