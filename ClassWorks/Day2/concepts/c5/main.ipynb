```python
import numpy as np
import matplotlib.pyplot as plt

def plot_moons_dataset(X, y):
    """
    Creates a scatter plot of the Moons dataset.
    
    Parameters:
    X: Feature matrix (n_samples, 2)
    y: Target labels (n_samples,)
    """
    plt.figure(figsize=(8, 6))
    plt.scatter(X[y == 0, 0], X[y == 0, 1], c='blue', label='Class 0', alpha=0.6, edgecolors='k')
    plt.scatter(X[y == 1, 0], X[y == 1, 1], c='darkred', label='Class 1', alpha=0.6, edgecolors='k')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title('Moons Dataset')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()

def plot_svm_decision_boundary(model, X, y, title):
    """
    Plots the decision boundary of an SVM model with the dataset.
    
    Parameters:
    model: Trained SVM classifier
    X: Feature matrix (n_samples, 2)
    y: Target labels (n_samples,)
    title: Title for the plot
    """
    plt.figure(figsize=(8, 6))
    
    # Create mesh grid
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                         np.linspace(y_min, y_max, 200))
    
    # Predict on mesh grid
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # Plot decision regions
    plt.contourf(xx, yy, Z, alpha=0.4, cmap='RdYlBu')
    
    # Plot data points
    plt.scatter(X[y == 0, 0], X[y == 0, 1], c='blue', label='Class 0', alpha=0.6, edgecolors='k')
    plt.scatter(X[y == 1, 0], X[y == 1, 1], c='darkred', label='Class 1', alpha=0.6, edgecolors='k')
    
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(title)
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.show()
```

```python
# Import Required Libraries
from sklearn.datasets import make_moons
from sklearn.svm import SVC
```

```python
# Generate the Moons dataset
X, y = make_moons(n_samples=300, noise=0.2, random_state=42)
```

```python
plot_moons_dataset(X, y)
```

```python
# Train Linear SVM
linear_svm = SVC(kernel='linear', random_state=42)
linear_svm.fit(X, y)
```

```python
plot_svm_decision_boundary(linear_svm, X, y, 'Linear SVM Decision Boundary')
```

```python
# Train Non-Linear SVM (RBF Kernel)
rbf_svm = SVC(kernel='rbf', random_state=42)
rbf_svm.fit(X, y)
```

```python
plot_svm_decision_boundary(rbf_svm, X, y, 'Non-Linear SVM (RBF) Decision Boundary')
```
