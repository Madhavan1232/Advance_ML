```python
import pandas as pd

url = "https://github.com/Madhavan1232/Advance_ML/raw/refs/heads/main/ClassWorks/Day2/p4/ML470_S2_Practice_HR_Data.xlsx"
df = pd.read_excel(url)
print(df.head())
```

```python

from sklearn.preprocessing import LabelEncoder

le_salary = LabelEncoder()
df['salary.enc'] = le_salary.fit_transform(df['salary'])

le_department = LabelEncoder()
df['Department.enc'] = le_department.fit_transform(df['Department'])

df = df.select_dtypes(include=['number'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True , cmap='YlGnBu')
```
